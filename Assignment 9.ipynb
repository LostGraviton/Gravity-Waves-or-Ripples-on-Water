# Water Waves :

### Only the effect gravity considered :

The wave function consists of sum of two Gauss functions : $ \psi (x,t) = \psi_1 (x,t) + \psi_2 (x,t)$

Where, $\psi_1(x,t) = \frac{a}{2\sqrt{\pi}} \int_{-\infty}^{+\infty} exp\left[ \frac{-(k+k_0)^2}{4} \right] \cos(kx - \sqrt{g|k|} \, t) \, dk $

And, $\psi_2(x,t) = \frac{a}{2\sqrt{\pi}} \int_{-\infty}^{+\infty} exp\left[ \frac{-(k-k_0)^2}{4} \right] \cos(kx - \sqrt{g|k|} \, t) \, dk $

So, now let's solve these two integral equations numerically :

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math as mth
import numpy as np
sns.set_style('darkgrid')
```

```python
def f1(x,k,t):
    return (a/mth.sqrt(4*np.pi)) * (mth.exp((-((k+k0)*a)**2)/4)) * (np.cos(k*x - np.sqrt(abs(g*k))*t))  # defined the function to integrate

def f2(x,k,t):
    return (a/mth.sqrt(4*np.pi)) * (mth.exp((-((k-k0)*a)**2)/4)) * (np.cos(k*x - np.sqrt(abs(g*k))*t))  # defined the function to integrate


# defined initial conditions

xl = -100.0
xr = 100.0
N = 2000
H = (xr-xl)/N # step size for x
a = 6
lam0 = 10.0
k0 = (2* np.pi)/lam0
g = -9.8
kl = -8.0/a
kr = 8.0/a
n = 1000
h = (kr-kl)/n # step size for k
x = xl
t = 0

# defined the lists

X1 = []
F1 = []
X2 = []
F2 = []
X3 = []
F3 = []


# Now let's integrate
    
for _ in range(N):
    
    integration1 = f1(x,kl,t) + f1(x,kr,t)

    for i in range(n):
        
        j1 = kl + i*h

        if i%2 == 0:
            integration1 = integration1 + 2 * f1(x,j1,t)  # for even terms
  
        else:
            integration1 = integration1 + 4 * f1(x,j1,t)  # for odd terms

    integration1 = integration1 * h/3
    
    integration2 = f2(x,kl,t) + f2(x,kr,t)

    for i in range(n):
        
        j2 = kl + i*h

        if i%2 == 0:
            integration2 = integration2 + 2 * f2(x,j2,t)  # for even terms
        else:
            integration2 = integration2 + 4 * f2(x,j2,t)  # for odd terms

    integration2 = integration2 * h/3   
    
    X1.append(x)
    F1.append(integration1 + integration2)
    x = x + H
    


# defined initial conditions

xl = -100.0
xr = 100.0
N = 2000
H = (xr-xl)/N # step size for x
a = 6
lam0 = 10.0
k0 = (2* np.pi)/lam0
g = -9.8
kl = -8.0/a
kr = 8.0/a
n = 1000
h = (kr-kl)/n # step size for k
x = xl
t = 10

for _ in range(N):
    
    integration1 = f1(x,kl,t) + f1(x,kr,t)

    for i in range(n):
        
        j1 = kl + i*h

        if i%2 == 0:
            integration1 = integration1 + 2 * f1(x,j1,t)  # for even terms
  
        else:
            integration1 = integration1 + 4 * f1(x,j1,t)  # for odd terms

    integration1 = integration1 * h/3
    
    integration2 = f2(x,kl,t) + f2(x,kr,t)

    for i in range(n):
        
        j2 = kl + i*h

        if i%2 == 0:
            integration2 = integration2 + 2 * f2(x,j2,t)  # for even terms
        else:
            integration2 = integration2 + 4 * f2(x,j2,t)  # for odd terms

    integration2 = integration2 * h/3   
    
    X2.append(x)
    F2.append(integration1 + integration2)
    x = x + H



# defined initial conditions

xl = -100.0
xr = 100.0
N = 2000
H = (xr-xl)/N # step size for x
a = 6
lam0 = 10.0
k0 = (2* np.pi)/lam0
g = -9.8
kl = -8.0/a
kr = 8.0/a
n = 1000
h = (kr-kl)/n # step size for k
x = xl
t = 20    

for _ in range(N):
    
    integration1 = f1(x,kl,t) + f1(x,kr,t)

    for i in range(n):
        
        j1 = kl + i*h

        if i%2 == 0:
            integration1 = integration1 + 2 * f1(x,j1,t)  # for even terms
  
        else:
            integration1 = integration1 + 4 * f1(x,j1,t)  # for odd terms

    integration1 = integration1 * h/3
    
    integration2 = f2(x,kl,t) + f2(x,kr,t)

    for i in range(n):
        
        j2 = kl + i*h

        if i%2 == 0:
            integration2 = integration2 + 2 * f2(x,j2,t)  # for even terms
        else:
            integration2 = integration2 + 4 * f2(x,j2,t)  # for odd terms

    integration2 = integration2 * h/3   
    
    X3.append(x)
    F3.append(integration1 + integration2)
    x = x + H     
```

```python
plt.subplot(311)
plt.title("Water Wave")
plt.plot(X1, F1, label = "t = 0", color = "red")    
plt.legend()
plt.rcParams["figure.figsize"] = (20, 10)
plt.xlabel("x")
plt.ylabel("F")

plt.subplot(312)    
plt.plot(X2, F2, label = "t = 10", color = "red" )    
plt.legend()
plt.rcParams["figure.figsize"] = (20, 10)
plt.xlabel("x")
plt.ylabel("F")

plt.subplot(313)    
plt.plot(X3, F3, label = "t = 20",color = "red" )    
plt.legend()
plt.rcParams["figure.figsize"] = (20, 10)
plt.xlabel("x")
plt.ylabel("F")
```

And, $\psi_2(x,t) = \frac{a}{2\sqrt{\pi}} \int_{-\infty}^{+\infty} exp\left[ \frac{-(k-k_0)^2}{4} \right] \cos(kx - \sqrt{g|k| - \frac{\gamma}{\rho}|k|^{3}} \, t) \, dk $
